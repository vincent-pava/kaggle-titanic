# Titanic Survival Prediction

## Imports

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
```

## Data loading and cleaning

```python
test = pd.read_csv('./test.csv')
train = pd.read_csv('train.csv')
```

Dropping useless columns, preping data...

```python
train = train.drop(['Ticket', 'Name'], axis=1)
missing_values = train.isnull().sum()
print(f'Number of entries : {train.size}')
print(f'Missing values : \n {missing_values}')
# As the number of missing values is small compared to the number of entries we can simply ignore them 
train = train.dropna()
train.head()
```

'Les femmes et les enfants d'abord' 

```python
plt.figure(figsize=[40,10])
plt.subplot(1, 3, 1)
sns.barplot(x=train.Sex, y=train.Survived, ci=None)

plt.subplot(1,3,2)
sns.barplot(x=train.Pclass, y=train.Survived, ci=None)

plt.subplot(1,3,3)
sns.barplot(x=train.Age, y=train.Survived, ci=None)

```
